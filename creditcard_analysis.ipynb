```python
x=data.iloc[:,:-1]

y=data["Class"]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split as tts

trainx,testx,trainy,testy=tts(x,y, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import GridSearchCV,KFold
```

```python
model=LogisticRegression()

para={"C": 10.0**np.arange(-2,3), "penalty": ["l1","l2"]}
cv=KFold(n_splits=5, shuffle=False, random_state=None)
```

```python
bestmodel=GridSearchCV(model,param_grid=para,cv=cv,n_jobs=-1,scoring="f1_macro")

bestmodel.fit(trainx,trainy)
```

```python
ypred=bestmodel.predict(testx)
```

```python
print(accuracy_score(testy,ypred)*100)

print(classification_report(testy,ypred))

print(confusion_matrix(testy,ypred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model1=RandomForestClassifier()
model1.fit(trainx,trainy)
```

```python
ypred1=model1.predict(testx)
```

```python
print(accuracy_score(testy,ypred1)*100)

print(classification_report(testy,ypred1))

print(confusion_matrix(testy,ypred1))
```

# OverSampling

```python
from imblearn.over_sampling import RandomOverSampler

os=RandomOverSampler(0.5)

trainxos,trainyos=os.fit_sample(trainx,trainy)
```

```python
from collections import Counter
print(Counter(trainy))

print(Counter(trainyos))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2=RandomForestClassifier()
model2.fit(trainxos,trainyos)
```

```python
ypred2=model2.predict(testx)


print(accuracy_score(testy,ypred2)*100)

print(classification_report(testy,ypred2))

print(confusion_matrix(testy,ypred2))
```

# SMOTETomek TECHNIQUE

```python
from imblearn.combine import SMOTETomek

sm=SMOTETomek(0.5)

trainxos,trainyos=sm.fit_sample(trainx,trainy)
```
